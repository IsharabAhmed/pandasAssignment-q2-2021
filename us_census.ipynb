```python
import numpy as np
import pandas as pd
```

```python
import glob

allFiles = glob.glob('./*csv')

li = []

for filename in allFiles:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)
    
us_census = pd.concat(li, axis=0, ignore_index=True)
us_census.head(10)
```

```python
us_census.dtypes
```

```python
us_census.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
us_census.dtypes
```

```python
us_census.head(10)
```

```python
us_census.drop(columns=['location'], inplace=True)
us_census.drop(columns=['product_type'], inplace=True)
us_census.drop(columns=['product_description'], inplace=True)
us_census.drop(columns=['quantity'], inplace=True)
us_census.drop(columns=['price'], inplace=True)
```

```python
us_census.head(10)
```

```python
us_census[['Male' , 'Female']] = us_census.GenderPop.str.split('_', expand=True)
```

```python
us_census.head(10)
```

```python
us_census.drop(columns=['GenderPop'], inplace=True)
```

```python
us_census.head(10)
```

```python
us_census = us_census.replace('F','', regex=True)
```

```python
us_census.head(10)
```

```python
us_census = us_census.replace('M','', regex=True)
```

```python
us_census.head(10)
```

```python
us_census = us_census.replace('%','', regex=True)
```

```python
us_census.head(10)
```

```python
us_census['Income'] = us_census.Income.str.strip('$')
```

```python
us_census.head(10)
```

```python
us_census.loc[:,'Hispanic':'Income'] = round(us_census.loc[:,'Hispanic':'Income'].apply(pd.to_numeric), 2)
```

```python
us_census.head(10)
```

```python
us_census.dtypes
```

```python
us_census['Male'] = us_census['Male'].astype(float)
```

```python
pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
us_census
```

```python
us_census.dropna(inplace=True)
```

```python
us_census.dtypes
```

```python
us_census['Female'] = us_census['TotalPop'] - us_census['Male']
```

```python
us_census.dtypes
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(us_census['Female'],us_census['Income'])
plt.xlabel('Salary $', fontsize=20)
plt.ylabel('Female Population', fontsize=18)
plt.show()
```

```python
us_census.duplicated().any()
```

```python
us_census.drop_duplicates(inplace=True)
```

```python
plt.scatter(us_census['Female'],us_census['Income'])
plt.xlabel('Salary $', fontsize=20)
plt.ylabel('Female Population', fontsize=18)
plt.show()
```
